```python
X=df.drop('quality',axis=1)
X
```

```python
y=df['quality']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=123,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
y_train_pred=dt.predict(X_train)
```

```python
y_test_pred=dt.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print('Train Data:\n',classification_report(y_train,y_train_pred))
print('Test Data:\n',classification_report(y_test,y_test_pred))
```

```python
param_grid={'criterion':['gini','entropy'],
            'max_depth':np.arange(1,20),
            'min_samples_split':np.arange(1,50),
            'min_samples_leaf':np.arange(1,50),
            'splitter':['best','random'],
            'max_features':np.arange(1,40),
            'max_leaf_nodes':np.arange(1,25) 
           }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_clf=GridSearchCV(estimator=dt,param_grid=param_grid,n_jobs=-1,cv=7,)
```

```python
grid_clf.fit(X_train,y_train)
```

```python
grid_clf.best_params_
```

```python
grid_clf.best_score_
```

```python
grid_train_pred=grid_clf.predict(X_train)
grid_test_pred=grid_clf.predict(X_test)
```

```python
print(classification_report(y_train,grid_train_pred))
print(classification_report(y_test,grid_test_pred))
```
